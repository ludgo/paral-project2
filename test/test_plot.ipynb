```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
```

```python
import sys
name = sys.argv[1]
```

```python
f = pd.read_excel('{}.xlsx'.format(name))
```

```python
n = np.arange(1,f.shape[0]+1)
```

```python
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 16
fig_size[1] = 16
plt.rcParams["figure.figsize"] = fig_size
```

```python
for col in f:
    plt.scatter(n, f[col], label='')
    line_style = '-' if col.startswith('t1') or col.startswith('static') else '--'
    plt.plot(n, f[col], linewidth=3, linestyle=line_style)

plt.xticks(n)
plt.ylim( (0, f.values.max() * 1.1) )

plt.xlabel('# of measurement')
plt.ylabel('time [s]')
plt.legend(loc='upper right', ncol=3)

plt.grid(b=True, which='major', linestyle='-', color='black')
plt.grid(b=True, which='minor', linestyle='--')
plt.minorticks_on()
```

```python
pdf_page = PdfPages('{}.pdf'.format(name))
plt.savefig(pdf_page, format='pdf')
pdf_page.close()
```
